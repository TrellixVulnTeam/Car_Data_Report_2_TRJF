```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data= pd.read_csv("data.csv",sep=",")
#print(data)
```

```python
missingtable= data.isnull()
#print(missingtable)

dict={}
count=0

for k in range(0,len(missingtable.columns)):
    dict[missingtable.columns[k]]=0

for i,j in missingtable.iterrows():
    for k in range(0,len(missingtable.columns)):
        if j[k]:  
            dict[missingtable.columns[k]]+=1
        
for k in range(0,len(missingtable.columns)):
    print(missingtable.columns[k]," ",dict[missingtable.columns[k]])
```

```python
enginehp={}

for i,j in data.iterrows():
    enginehp[j[0]+" "+j[1]+" "+str(j[2])]=0

for i,j in missingtable.iterrows():
    for k in range(0,len(missingtable.columns)):
        if j[k] and k==4: enginehp[data.loc[i][0]+" "+data.loc[i][1]+" "+str(data.loc[i][2])]+=1

for key in enginehp:
    if enginehp[key]: print() #print(key,'->',enginehp[key])
    
```
